## Dataset Description

This dataset comes from NYC Open Data and can be accessed through [Evictions](https://data.cityofnewyork.us/City-Government/Evictions/6z8x-wfk4/data). This dataset lists pending, scheduled and executed evictions within the five boroughs, for the year 2017 - Present. The data have fields including Court Index Number, Docket Number, Eviction Address, Apartment Number, Executed Date, Marshal First Name, Marshal Last Name, Residential or Commercial (property type), Borough, Zip Code, Scheduled Status (Pending/Scheduled) and so on. Eviction data is compiled from the majority of New York City Marshals. Marshals are independent public officials and should be contacted directly for more information at https://www1.nyc.gov/site/doi/offices/marshals-list.page Data prior to January 1 2017 is not currently available.

## Ratio of Commercial to Residential Evictions

What is the ratio of commercial to residential evictions in the entire data? Was this ratio constant over time (calculate it separately for each year in the data)? Visualize the data for each year and discuss the trends you found.

### Ratio in entire data

```python
evictions['Residential/Commercial'].value_counts()
```

```python
#number of residential evictions 
residential = evictions[evictions['Residential/Commercial']=='Residential']['Court Index Number'].count()
#number of commercial evictions
commercial = evictions[evictions['Residential/Commercial']=='Commercial']['Court Index Number'].count()
#calculate the ratio
ratio = commercial/residential
print('Ratio of Commercial : Residential is 1 :', ratio)
```

### Trend over time (in each year)

```python
#define a function to calculate the ratio in each year
def ratio_year(year):
    evictions_year= evictions[(evictions['year']==year)] #filter for sepcific year
    residential = evictions_year[evictions_year['Residential/Commercial']=='Residential']['Court Index Number'].count()
    commercial = evictions_year[evictions_year['Residential/Commercial']=='Commercial']['Court Index Number'].count()
    ratio = commercial/residential
    print(year, ': Ratio of Residential : Commercial is 1 :', ratio)
    return ratio
```

```python
ratio_year(2017)
```

```python
#group by year and commercial/residential columns and count the number of each
ratio_trend = evictions.groupby(['year','Residential/Commercial'])['Court Index Number'].agg(['count']).unstack()

#add a ratio column
ratio_trend['ratio'] = ratio_trend['count']['Commercial']/ratio_trend['count']['Residential']

#drop incompleted year/outlier
ratio_trend = ratio_trend.drop([2021, 2070])
ratio_trend
```

### Plot the trend

```python
#plot the line graph
ratio_trend.plot.line(y='ratio', figsize=(8,6),
                     title = 'Ratio of Commercial to Residential Evictions by year',
                     xlabel = 'Year',
                     ylabel = 'Ratio',
                     legend = False)
plt.locator_params(axis="x", nbins=4)
```

The ratio of commercial to residential evictions is increasing from 2017 to 2020, which means that number of residential evitions decreased is larger than number of commercial eviction decreased. Residential evictions decreased more than commercial evictions in past 4 years. The trend as we can see from the graph shows that they are positively related and the gradient is quite constant, which means the ratio is constantly increasing over time.

## Compare 5 Boroughs

How do 5 boroughs compare overtime in terms of the number of evictions in each? Discuss the findings and compare them to each boroughâ€™s population counts.

### Number of evictions in 5 boroughs over time

```python
#group by borough and year and count the number of rows
#filter out incompleted year/outlier
borough = evictions[evictions['year']<2021].groupby(['BOROUGH','year'])['Court Index Number'].agg(['count'])
borough.head()
```

```python
#plot to compare between boroughs over time
borough['count'].unstack().plot.bar(figsize=(8,6),
                                    title = 'Number of Evictions in 5 boroughs from year 2017-2020',
                                    xlabel = 'Borough',
                                    ylabel = 'Number of Evictions')
plt.xticks(rotation = 0)
```

From the above graph and table, we could see that Bronx has the highest number of evictions in all four past years, followed by Broolyn and Queens. Almost all the boroughs have decreasing trend of number of evictions over time from 2017 to 2020, except for Queens which has a slightly increase from 2017 to 2018. As simply comparing number of evictions without considering the population of each borough will lead to some bias, we would divide the number of evictions by population in the next section for further comparsion.

## Visualize evictions in NTA level

Visualize the data as a choropleth map using NTA level, when the different shades represent the number of evictions in a particular NTA. Discuss your findings.

The Neighborhood Tabulation Area dataset comes from NYC Open Data and can be accessed through [NTA](https://data.cityofnewyork.us/City-Government/Neighborhood-Tabulation-Areas-NTA-/cpf4-rkhq).

```python
nta = gpd.read_file('Neighborhood Tabulation Areas (NTA)')
```

```python
#group by NTA and count number of evictions/rows per NTA
nta_evictions = evictions.groupby('NTA')['Court Index Number'].agg(['count']).reset_index()

#see the statistics of the evictions count in NTA
nta_evictions['count'].describe()
```

```python
#merge two datasets
nta_evictions_plot =  nta.merge(nta_evictions, left_on='ntaname', right_on='NTA')
```

```python
#plot the graph

#set the range for the choropleth
vmin, vmax = 6,1363

#create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(20, 10))

#remove the axis
ax.axis('off')

#add a title and annotatio"
ax.set_title("Number of Evictions by NTA in NYC", fontdict={'fontsize': '20', 'fontweight' : '3'})

#Create colorbar legend
sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))

#empty array for the data range
sm.set_array([])

#add the colorbar to the figure
fig.colorbar(sm, shrink=0.5)

#create an annotation for the data source
ax.annotate('Source: NYC Open Data',xy=(0.7, 0.1),  xycoords='figure fraction', 
            horizontalalignment='left', verticalalignment='top', fontsize=12, color='#555555')

#plot the map
nta_evictions_plot.plot(column= 'count', cmap='Reds', linewidth=0.2, ax=ax, edgecolor='0.8')
```

```python
#sort the value to see the top 10 NTAs that have most evictions
nta_evictions_plot.sort_values(by='count',ascending=False).tail(10)
```

From the table and map above, we could see that neighborhood tabulation areas in Bronx and Brooklyn have the highest number of evictions in past four years, especially in East New York, Crown Heights North, and East Concourse-Concourse Village that have over 1000 evictions. Areas of park-cemetery-etc have the lowest number of evictions which is easy to understand since there are not too many housing in these areas. Many areas in upper Queens and Staten Islands have less evictions overall. 

## Socio-demographic attribute that affects evictions
### Median Household Income

Can you think of one additional socio-demographic attribute that may affect the number of evictions in a particular neighborhood? Fetch data on that attribute and visualize it side-by-side with evictions or in the same map. Discuss your assumptions and findings as well as possible connections.

The Median Household Income dataset comes from Pratt neighborhood data protal and can be accessed through [MHI](https://prattcenter.net/resources/neighborhood_data_portal).

```python
#import the median household income shapefile
median_income = gpd.read_file('Median Household Income')

#median_income.head()
```

The median household income has the geometry in census tract level. In order to transfer it into NTA level for comparison to the evictions map, I used the census tract shapefile to add the nta attribute. The Census Tract dataset comes from NYC Open Data and can be accessed through [census tract](https://data.cityofnewyork.us/City-Government/2010-Census-Tracts/fxpq-c8ku).

```python
#import the census tract shapefile for joining as it contains nta attribute
census_tract = gpd.read_file('2010 Census Tracts')

#census_tract.head()
```

```python
#merge the census tract and median household income datasets to join the nta attribute
cs_median = census_tract.merge(median_income, left_on='ct2010', right_on='TRACT')

#cs_median.head()
```

```python
#group by nta and get the mean of MHI in several smaller census tracts for larger nta
nta_median = cs_median.groupby('ntacode')['MHI_2'].agg(['mean']).reset_index()

#get the statistics of MHI in nta level
nta_median['mean'].describe()
```

```python
#merge the MHI in nta level and nta shapefile
nta_median_plot =  nta.merge(nta_median, left_on='ntacode', right_on='ntacode')
nta_median_plot.head()
```

```python
## Plot the median household income by NTA map

#set the range for the choropleth
vmin1, vmax1 = 25914,140948

#create figure and two axes/columns for side by side map
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

#remove the axis
ax1.axis('off')

#add a title and annotatio"
ax1.set_title("Median Household Income by NTA in NYC (US Dollar)", fontdict={'fontsize': '20', 'fontweight' : '3'})

#Create colorbar legend
sm1 = plt.cm.ScalarMappable(cmap='Oranges', norm=plt.Normalize(vmin=vmin1, vmax=vmax1))

#empty array for the data range
sm1.set_array([])

#add the colorbar to the figure
fig.colorbar(sm1, ax = ax1, shrink=0.5)

#create an annotation for the data source
ax1.annotate('Source: NYC Open Data, Pratt Neighborhood Data Portal',xy=(0.65, 0.12),  xycoords='figure fraction', 
            horizontalalignment='left', verticalalignment='top', fontsize=12, color='#555555')

nta_median_plot.plot(column= 'mean', cmap='Oranges', linewidth=0.2, ax=ax1, edgecolor='0.8')



## Plot the evictions by NTA map

#set the range for the choropleth
vmin, vmax = 6,1363

#remove the axis
ax2.axis('off')

#add a title and annotatio"
ax2.set_title("Number of Evictions by NTA in NYC", fontdict={'fontsize': '20', 'fontweight' : '3'})

#Create colorbar legend
sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))

#empty array for the data range
sm.set_array([])

#add the colorbar to the figure
fig.colorbar(sm, ax = ax2, shrink=0.5)

nta_evictions_plot.plot(column= 'count', cmap='Reds', linewidth=0.2, ax=ax2, edgecolor='0.8')
```

I choose median household income as one additional socio-demographic attribute that may affect the number of evictions in particular neighborhoods, because I think neighborhood areas with higer median household income would be more likely be able to pay the rents and are less likely to be evicted because of rent issues. The result in the side by side map above quite fits my assumption as areas with higher number of evictions (darker red) on the right has lower median household income (lighter orange) on the left and vice versa. Particulalry in neighborhoods such as lower Manhattan, eastern Brooklyn and Bronx where they show a more stirking contrast in median household income and number of evictions. 

I've also noticed that the median household income are quite average across several neighborhoods except some areas but the evictions are relatively concentrated on small number of areas in Brooklyn and Bronx. Overall, there is a direct connection between median household income and number of evitions in different neighborhood tabulation areas.
